```python
#a) Take dynamic input from user as a paragraph (give input >= 25 lines) and remove punctuations first (only). 
#Then print resulting paragraph (after removing punctuations) with all lower case 

#Take Input >=25 Lines
Dynamic_input=input("ENTER THE INPUT >=25 LINES: \n")
print("The given Input is :\n"+Dynamic_input)

#remove_punctuations
exclude = set(string.punctuation)
s = ''.join(ch for ch in Dynamic_input if ch not in exclude)
print("\n\n PUNCUATION REMOVED TEXT: \n"+ s)
s=s.lower()
print("\n\n PUNCUATION REMOVED TEXT AND LOWER CASED: \n"+ s)

```

```python
#b) Afterwards, remove stop words and print the whole paragraph again.

import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
  
example_sent = s
  
stop_words = set(stopwords.words('english')) 
  
word_tokens = word_tokenize(example_sent) 
  
filtered_sentence = [w for w in word_tokens if not w in stop_words] 
  
filtered_sentence = [] 
  
for w in word_tokens: 
    if w not in stop_words: 
        filtered_sentence.append(w) 
print(word_tokens) 
print(filtered_sentence) 
```

```python
# c) Collect those words that occur only once in the paragraph and print them.

def word_count1(slc):
    counts = dict()
    words = slc

    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1
    return counts

sdwe=word_count1(filtered_sentence)
print(sdwe)

def getKeysByValue(dictOfElements, valueToFind):
    listOfKeys = list()
    listOfItems = dictOfElements.items()
    for item  in listOfItems:
        if item[1] == valueToFind:
            listOfKeys.append(item[0])
    return  listOfKeys


listOfKeys = getKeysByValue(sdwe, 1)
print("\nKeys with word count value equal to 1")
#Iterate over the list of keys
for key  in listOfKeys:
    print(key)
```

```python
# d) Collect those words that occur only twice in the paragraph and print them. 
listOfKeys = getKeysByValue(sdwe, 2)
print("\nKeys with word count value equal to 2")
#Iterate over the list of keys
for key  in listOfKeys:
    print(key)
```

```python
# e) Make a dictionary of these 4 types of words and apply stemming to reduce inflected words to their word stem,
# base or root form. 


# import these modules 
from nltk.stem import PorterStemmer 
from nltk.tokenize import word_tokenize 

ps = PorterStemmer() 

# choose some words to be stemmed 
words = filtered_sentence

for w in words: 
	print(w, " : ", ps.stem(w)) 
```
